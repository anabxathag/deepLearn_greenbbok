# (1.) Perceptron อย่างง่าย

image.png

s คือ ผลรวมหรือ Sum = bias(w0) + [ค่า Input แต่ละตัว(xi) คูณกับ weight(wi) ...]
-   s = w0 + (w1 * x1) + (w2 * x2) + ... + (wn * xn)



จาเเรื่อง activation function;
-   s >= t , y = 1
-   s < t, y = 0

โดย t(Thereshold, ค่าเกณฑ์) จะกำหนดเป้นเท่าใดก็ได้ ซึ่งในหัวข้อนี้ให้ t = 0


```python
def activa_func(s):
  t = 0
  if s >= t:
    return 1
  else:
    return 0
```

**ex.1**

```python
import numpy as np
w0 = -.5 # bias = -0.5
x = [[1],
     [1]] # input x1, x2 (เขียนบรรทัดเดียวกันก็ได้)
w = [.5, .5]  # weight w1, w2

w, x = np.array(w), np.array(x) # เปลี่ยนเป็น numpy array

s1 = w0 + np.dot(w,x)  # คำนวณด้วย Dot Product
y1 = activa_func(s1) # ค่า y(ผลลัพธ์)
print(s1, '-->', y1)
```

s = 0.5 --> y = 1

**ex.2**

```python
w0 = -.5 # bias = -0.5
x = [[0],
     [0]] # input x1, x2 (เขียนบรรทัดเดียวกันก็ได้)
w = [.5, .5]  # weight w1, w2

w, x = np.array(w), np.array(x) # เปลี่ยนเป็น numpy array

s2 = w0 + np.dot(w,x)  # คำนวณด้วย Dot Product
y2 = activa_func(s2) # ค่า y(ผลลัพธ์)
print(s2, '-->', y2)
```

s = -0.5 --> y = 0

# (3.) AND Gate Neuron

image.png

```python
def and_gate(x1, x2, wei):
  w0 = -0.5   # bias
  w1 = w2 = wei   # weight
  t = 0.0  # thereshold
  s = w0 + (w1 * x1) + (w2 * x2)  # sum

  if s >= t:  # ตัดสินโดยเทียบกับ thereshold
    result = 1
  else:
    result = 0

  print('output y={}  [s={}]'.format(result, s))
```

ใช้ค่า weight = 0.5

```python
and_gate(0, 0, 0.5)   # output ถูก
and_gate(0, 1, 0.5)   # output ผิด
and_gate(0, 1, 0.5)   # output ผิด
and_gate(1, 1, 0.5)   # output ถูก
```

จากผลลัพธ์หรือค่า y ที่ได้ เมื่อเทียบกับ target(ตาราง AND Gate) --> ถูก 2 เคส, ผิด 2 เคส

ต้องทำ Backpropagation คือ การปรับค่า weight(ปรับ w0, w1, w2 ตัวใดตัวหนึ่ง หรือ ทั้งหมดก็ได้)

ลองใช้ weight = 0.2

```python
and_gate(0, 0, 0.2)   # output ถูก
and_gate(0, 1, 0.2)   # output ถูก
and_gate(0, 1, 0.2)   # output ถูก
and_gate(1, 1, 0.2)   # output ผิด
```

ถูก 3 เคส, ผิด 1 เคส --> การปรับค่า weight ยังไม่ได้ที่

ลองใช้ weight = 0.45

```python
and_gate(0, 0, 0.45)   # output ถูก
and_gate(0, 1, 0.45)   # output ถูก
and_gate(0, 1, 0.45)   # output ถูก
and_gate(1, 1, 0.45)   # output ถูก
```

เมื่อเทียบกับ target จะเห็นวาสถูกต้องทั้งหมดทุกเคส ไม่มี error แล้ว ไม่ต้องย้อนกลับไปปรับค่า weight อีก เท่ากับว่า Neuron หรือ Perceptron นี้ได้ทำการ Train เรียบร้อบแล้ว

# (4.) OR Gate Neuron

image.png

```python
def or_gate(x1, x2):
  w0 = -0.5   # bias
  w1 = 0.5   # weight ทดลองปรับค่าเองได้...
  w2 = 0.5
  t = 0.0  # thereshold
  s = w0 + (w1 * x1) + (w2 * x2)  # sum

  if s >= t:  # ตัดสินโดยเทียบกับ thereshold
    result = 1
  else:
    result = 0

  print('output y={}  [s={}]'.format(result, s))
```

```python
# ทำนายยยย
or_gate(0, 0)   # output ถูก
or_gate(0, 1)   # output ถูก
or_gate(0, 1)   # output ถูก
or_gate(1, 1)   # output ถูก
```

#(6.) Neuron จำแนกมะนาว

ทำNeuron ระบบจำแนก ว่าเป็นมะนาวหรือไม่ โดยมี Input รับค่า Feature น้ำหนัก(m) และขนาดเส้นผ่านศูนย์กลาง(d) (เทียบได้กับ Input x1 x2 ที่ผ่านมา) มี Output target (y) ดังผลลัพธ์ของโค้ดข้างล่างนี้

```python
def is_lemon(x1, x2):
  w0 = -3.5
  w1 = 0.2
  w2 = 0.5
  t = 0.0
  s = w0 + (w1 * x1) + (w2 * x2)

  if s >= t:
    result = 1
  else:
    result = 0

  print('output y={}  [s={}]'.format(result, s))

is_lemon(2, 4.2)
is_lemon(3, 2)
is_lemon(8, 3.1)
is_lemon(8, 4.2)  # ถ้ามี นน(m) และ ขนาด(d) ตามบรรทัดนี้ คือ มะนาว
```

จาก output ข้างบน พบว่า No Error

ที่นี้ ลองให้เป็น ส้มเขียวหวาน ที่มี m=8, d=6.5

```python
is_lemon(8, 6.5)
```

ได้ output = 1 นั่นคือ ระบบบอกว่า ส้มเขียวหวาน คือ มะนาว

แม้ในการคำนวณจะถูกต้อง แต่ไม่ถูกต้องในความจริง จึงเกิดปัญหาคล้ายกับ XOR Gate

# (7.) Binary Step Activation Function

มีลักษณะเป็นแบบขั้นบันได 2 ค่า
-   f(x) = 1, x >= 0
-   f(x) = 0, x < 0



```python
def binary_step(x):
  result = 1 if x > 0 else 0
  return result
```

```python
print(binary_step(0))      # f(0)
print(binary_step(-0.5))  # f(-0.5)
print(binary_step(0.4))   # f(0.4)
print(binary_step(2.5))   # f(2.5)
```

ทดลองสร้าง input หลายๆค่า โดยใช้ array

```python
import numpy as np

input = np.arange(-7, 7, 0.15)  # สร่าง input ช่วง -7 ถึง 6.8 Step ละ 0.15

# ลองดูข้อมูล10ตัวแรก และ10ตัวท้าย
print(input[:10])
print(input[-11:-1])
```

ลองส่งค่าเข้า input ประมวลผลแล้วดู output

```python
output = [binary_step(x) for x in input]

output[42:54]
```

plot ดูความสัมพันธ์input และ output(ดูคุณลักษณะของ Activation Function)

```python
import matplotlib.pyplot as plt

plt.plot(input, output)
plt.ylabel('y(Output)')
plt.xlabel('x(Input)')

plt.show()
```

# (8.) Linear Activation Funtion

มีลักษณะการทำงานเป็นแบบเชิงเส้น ตามสมการ **y = mx**

- y เป็น output ของ function
- x เป็น input ที่ป้อนเข้า function
- m เป็นค่าความชันของเส้นตรง

```python
def linear05(x):      # ลดความแรง
  result = 0.5 * x
  return result

def linear1(x):      # เท่าเดิม
  result = x
  return result

def linear2(x):      # เพิ่มเป็น 2 เท่า
  result = 2 * x
  return result
```

```python
print(linear05(3))  # 1.5
print(linear1(3))  # 3
print(linear2(3))  # 6
```

```python
input = np.arange(-7, 7, 0.15)
output05 = [linear05(x) for x in input]
output1 = [linear1(x) for x in input]
output2 = [linear2(x) for x in input]

plt.plot(input, output05, lw=2., c='g')
plt.plot(input, output1, lw=2., c='m')
plt.plot(input, output2, lw=2., c='b')
plt.ylabel('y(Output)')
plt.xlabel('x(Input)')
plt.axhline(y=0.0, color='r', linestyle='--', linewidth=1.2)
plt.grid()
plt.show()
```

# (9.) Sigmoid Function

S(x) = 1 / (1 + e^(-x)), 0 < S(x) < 1

e = ค่าคงที่ของ Euler's number

```python
from math import exp

def sigmoid(x):
  return 1 / (1 + exp(-x))

print(sigmoid(0))
print(sigmoid(1))
print(sigmoid(6))
print(sigmoid(50))
print(sigmoid(-8))
```

```python
input = np.arange(-7, 7, 0.2)
output = [sigmoid(x) for x in input]

plt.plot(input, output, c='b')
plt.ylabel('y(Output)')
plt.xlabel('x(Input)')
plt.axhline(y=0.5, color='r', linestyle='--', linewidth=1.2)
plt.grid()
plt.show()
```

# (10.) TanH Function

tanh(x) = ((e^x) - (e^-x)) / ((e^x) + e^(-x)), -1 < tanh(x) < 1

```python
import numpy as np
from math import exp

def tanh(x):
  return (exp(x) - exp(-x)) / (exp(x) + exp(-x))

print(tanh(0))
print(tanh(1))
print(tanh(10))
print(tanh(-1))
```

```python
input = np.arange(-6, 6, 0.2)
output = [tanh(x) for x in input]

plt.plot(input, output, c='g')
plt.ylabel('y(Output)')
plt.xlabel('x(Input)')
plt.axhline(y=0, color='r', linestyle='--', linewidth=1.2)
plt.grid()
plt.show()
```

# (11.) ReLU Function

-   f(x) = 0, x < 0
-   f(x) = x, otherwise

```python
def ReLu(x):
  return max(0, x)

print(ReLu(-1))
print(ReLu(-2))
print(ReLu(0.5))
print(ReLu(2))
```

```python
input = np.arange(-6, 6, 0.2)
output = [ReLu(x) for x in input]

plt.plot(input, output, c='b')
plt.ylabel('y(Output)')
plt.xlabel('x(Input)')
plt.grid()
plt.show()
```

# (12.) Leaky ReLu

-   f(x) = 0.01(x), x < 0
-   f(x) = x, otherwise

```python
def leaky_ReLu(x):
  if x < 0:
    return 0.01 * x
  else:
     return x

print(leaky_ReLu(-1))
print(leaky_ReLu(-2))
print(leaky_ReLu(0.5))
print(leaky_ReLu(2))
```

```python
input = np.arange(-16, 6, 0.2)
output = [leaky_ReLu(x) for x in input]

plt.plot(input, output, c='b')
plt.ylabel('y(Output)')
plt.xlabel('x(Input)')
plt.grid()
plt.show()
```

# (13.) Softmax Function

[Softmax Function คืออะไร](https://medium.com/super-ai-engineer/softmax-function-%E0%B8%84%E0%B8%B7%E0%B8%AD%E0%B8%AD%E0%B8%B0%E0%B9%84%E0%B8%A3-eae1f1bbef63)

```python
from math import exp
import numpy as np

def softmax(out):
  res = []
  esum = 0
  for z in out:
    esum += exp(z)
  for z in out:
    e = exp(z)
    res.append(e / esum)
  return np.array(res)

output = np.array([0.6, 1.9, 0.01])
result = softmax(output)
print(result)
```
